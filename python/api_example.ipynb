```python
import pandas as pd
import requests
import json
```

```python
# here is an example where we query the data between two dates and set the request limit to 10,000 rows (max is 50,000 rows)
url = "https://data.cityofnewyork.us/resource/76xm-jjuj?$where=incident_datetime between '2005-01-10T00:00:00' and '2005-01-20T00:00:00'&$limit=50000"
```

```python
r = requests.get(url)
```

```python
df = pd.DataFrame(json.loads(r.text))
```
